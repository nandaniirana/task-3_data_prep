```python
import pandas as pd

# Load your existing data
df = pd.read_csv('cleaned_data.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Task-3 specific fields
df['OrderMonth'] = df['Order Date'].dt.to_period('M').astype(str)
df['Year'] = df['Order Date'].dt.year
```

```python
# Value Segment (for deep-dive)
def value_segment(amount):
    if amount >= 6000: return 'High Value'
    elif amount >= 3000: return 'Medium Value'
    else: return 'Low Value'
df['ValueSegment'] = df['Amount'].apply(value_segment)
```

```python
# Customer Type (New vs Repeat)
customer_orders = df.groupby('CustomerName').size()
repeat_customers = customer_orders[customer_orders >= 2].index
df['CustomerType'] = df['CustomerName'].apply(
    lambda x: 'Repeat' if x in repeat_customers else 'New'
)
```

```python
# Save for Power BI
df.to_csv('task3_dashboard_data.csv', index=False)
print("✅ task3_dashboard_data.csv ready for Power BI!")
print("\nSample data:")
print(df[['OrderMonth', 'ValueSegment', 'CustomerType', 'Amount', 'Profit', 'Category', 'State']].head())
```
